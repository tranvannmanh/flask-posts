```python
import numpy as np
import sqlite3

con = sqlite3.connect('../../instance/recommend_system.sqlite')
db_env = con.cursor()
```

```python
db_env.execute('SELECT user_id, post_id FROM YOU_LIKE')
users_like_items = db_env.fetchall()
print(users_like_items)
```

```python
users_id = list(set([rec[0] for rec in users_like_items]))
posts_id = list(set([rec[1] for rec in users_like_items]))
# users_id.sort()
# posts_id.sort()
print('unique users_id: ', users_id)
print('unique posts_id: ', posts_id)
```

```python
rows_user_matrix = {}
for user_id in users_id:
  rows_user_matrix[user_id] = []
  for post_id in posts_id:
    if (user_id, post_id) in users_like_items:
      rows_user_matrix[user_id].append(1)
    else:
      rows_user_matrix[user_id].append(0)

rows_user_matrix
```

```python
list(rows_user_matrix.keys())[0]
```

```python
matrix_users_like_items = np.array([rows_user_matrix[user_id] for user_id in users_id])
print(matrix_users_like_items)
```
